```python
# 문자열의 길이를 구하는 len()함수가 있다.
# 입력된 주민번호 글자수를 알아내는 코드
# id/pw에 대한 글자수 검증용으로 활용할 수 있다.

inputSsn = "001010-1234567"

type(inputSsn)
```

```python
len(inputSsn)
```

```python
if len(inputSsn)==14:
    print("주민번호 14글자가 잘 입력 되었습니다.") # 참 True
else:
    print("주민번호 14글자가 잘못 입력 되었습니다.") # 거짓 False
```

```python
# 문자열 인덱싱은 문자를 표로 만들어 0번부터 주소를 넣은 값
# 문자열 슬라이싱은 문자를 인덱스로 잘라 가져옴 !
word = "Life is too short, You need Python"
#       0123456789                       33
```

```python
word[3] # word변수에 3번째 문자열
```

```python
word[-3] # 인데스에 -값은 뒤에서부터 ~
```

```python
# 변수[시작:-1끝] 끝 번호에는 해당 문자를 포함하지 않음.
# word[0:3] -> 0 <= a < 3
word[0:3]
```

```python
# 주민번호 1~2 자리를 생년으로 추출 year
# 주민번호 3~4 자리를 생월로 추출 month
# 주민번호 5~6 자리를 생일로 추출 day
# 주민번호 7번째 1자리를 남여 식별용 추출 sex


year = inputSsn[:2]
month = inputSsn[2:4]
day = inputSsn[4:6]
print("귀하에 생년은 " + year + "년")
print("귀하에 생월은 " + month + "월")
print("귀하에 생일은 " + day + "일 입니다.")

# 주민번호를 이용해서 남여 구분을 위한 코드
# 문자열로 인식된 숫자를 숫자로 계산시 타입변환이 필요
ssNumber = inputSsn[7]
# print(ssNumber)
# type(ssNumber)
ssNumber = int(ssNumber) # 문자가 숫자로 타입변환진행 int(문자)
# type(ssNumber)

if ssNumber % 2 == 1: # 숫자로 된 성별을 2로 나눈 나머지가 1
    sex = "남자"
else:
    sex = "여자"

print(sex)
```

```python
# 미션 주민번호 8번째 글자가 1, 2, 5, 6이면 1900년 생
# 3, 4, 7, 8 이면 2000년생으로 만들어 보세요
# inputSsn = "001010-1234567"

if ssNumber == 1 :
    yearif = "19"+year
elif ssNumber == 2 :
    yearif = "19"+year
elif ssNumber == 5 :
    yearif = "19"+year
elif ssNumber == 6 :
    yearif = "19"+year
elif ssNumber == 9 :
    print("당신은 외계인입니다.")
else:
    yearif = "20"+year

print(yearif)
```
