```python
# 리스트란 표 형식의 배열을 말한다.
# ids = ["kkw","jhj","ksb"]
# pws = ["1234","4321","4567"]
# names = ["김기원","임효정","김수빈"]
# 리스트를 만들 때는 대괄호[] 로 감싸주고 쉼표, 로 구분한다.

# a = [] 빈 리스트를 생성한다. -> a = list()
# b = [1, 2, 3] 정수를 입력 할 수 있다. -> b[1] => 2
# c = [1, 2, 'kkw', 'lhk']와 같이 혼용하여 넣을 수 있다.
#     학년 반  이름     id
# d = [1, 2, ['kkw', '1234']] 2차원 리스트
#               0       1
#      0  1         2
# d 변수에서 kkw를 가져오고 싶으면?? d[2][0]
# f = [1, 2, ['kkw', '1234', ['파이썬', '자바']]] 3차원 리스트
#                                0        1
#              0       1            2
#      0  1                 2
# f 변수에서 파이썬을 가져오고 싶으면?? f[2][2][0]
```

```python
# 리스트에 C R U D 기업을 적용해보자.
# CREATE -> 생성, 추가, 삽입 .append()
# READ   -> 전체리스트보기, 1개 자료 보기 .index()
# UPDATE -> 자료 수정 passwords[3] = "4486"
# DELETE -> 자료 삭제 del passwords[3] .remove(3) .pop(3)
```

```python
# 리스트에 자료를 추가해보자!!!
ids = ["kkw","lhj","ksb"] # 회원 id
ids.append("mbc")
print(ids)
```

```python
passwords = ["1234","4321","4567"] # 비밀번호 password
passwords.append("8888")
print(passwords)
```

```python
# 리스트에 있는 값 가져오기
ids.index("mbc")
```

```python
# mbc id에 인덱스를 알아봐 패스워드의 값을 찾아온다.
passwords[3]
```

```python
# 알아온 password 값을 4486으로 수정해본다.
passwords[3] = "4486"
print(passwords)
```

```python
# 3번째 인덱스에 암호를 삭제해본다.
passwords.remove("4486")
print(passwords)
```

```python
# pop은 삭제되는 내용을 출력하고 삭제한다.
# passwords.pop() -> 맨 마지막 자료를 삭제
# passwords.pop(2) -> 2 인덱스를 삭제한다.
passwords.pop(2)
```
