```python
# if문은 조건문으로 판단하여 참(True) 거짓(False)로 판단하는 코드
# if 조건 :
#    조건이 참일 때 실행하는 문
# elif 조건 :
#    위 조건이 false일때 또 비교하는 조건문
# else:
#    위 조건이 모두 해당하지 않을때 실행하는 조건문

# 학점 계산용 코드
avg = int(input("평균을 입력하세요 >>>"))
if avg >= 90 :
    grade = "A" # 90점 이상이면 A

elif avg >= 80 : # 80점 이상이면 B
    grade = "B"

elif avg >= 70 : # 70점 이상이면 C
    grade = "C"

else :
    grade = "F" # 나머지

print("귀하의 등급은 : " + grade + "입니다.")
```

```python
marks = [90, 25, 67, 45, 80]
number = 0
for mark in marks :
    number = number +1
    if mark > 60 :
        print(f"{number}번 학생 합격입니다.")
    else:
        print(f"{number}번 학생 불합격입니다.")
```

```python
marks = [90, 25, 67, 45, 80]
number = 0
for mark in marks:
    number = number +1
    if mark < 60 :
        continue
    print(f"{number}번 학생 합격입니다.")
```
