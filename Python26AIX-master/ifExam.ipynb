```python


# if문은 조건문으로 판단하여 참(True) 거짓(False)로 판단하는 코드
# if 조건 :
#    조건이 참일 때 실행하는 문
# elif 조건 :
#    위 조건이 false일때 또 비교하는 조건문
# else :
#    위 조건이 모두 해당하지 않을때 실행하는 조건문

# 학점 계산용 코드
avg = int(input("평균을 입력하세요 >>>"))
if avg >= 90 :
    grade = "A" # 90점이상이면 A

elif avg >= 80 :
    grade = "B" # 80점이상이면 B

elif avg >= 70 :
    grade = "C" # 70점이상이면 C

else:
    grade = "F"

print("귀하의 등급은 : " + grade + "입니다.")
```
