```python
# 함수는 반복적인 코드를 위쪽에 선언하여 호출함으로써
# 프로그램을 가독성 좋고 간편하게 구현할수 있다.

# 클래스가 없는 기능을 함수라고 부르고
# 클래스가 있는 기능을 메서드라고 한다.

# def 는 함수나 메서드를 만드는 예약어로
# def 함수명(매개변수 == 인수 == 파라미터) :
#                                  들여쓰기 필수
#    함수내부의 실행문들.....
#    return 값  함수종료시 호출한 곳으로 넘겨주는 값
# 들여쓰기를 끝내면 함수 종료

# 변수 = 함수명(매개변수 == 인수 == 파라미터)
# 위에서 만든 함수를 호출하여 처리후 리턴을 제공 받는다.
```

```python
# 함수를 만든다. (함수 정의)
def add(a, b): # add() 함수를 만들고 인수는 a와 b를 받는다.
    return a + b    # 받은 a, b를 더하고 돌려준다.
```

```python
# 위에서 만든 함수를 활용해보자.
c = add(5,10) # 위에서 만든 add 함수에 5와 10을 넣고 결과를 c에 넣음
print(c)
```

```python
# 만약 입력값이 몇개가 될지 모를때는??
def add_many(*args): # *을 붙이면 몇개의 값이 들어오든!!!
    # *의 뜻은 와일드 문자로 김씨로 시작한다.~ "김*"
    # "*원" 원으로 끝나는 문자
    # "*딸기*" 딸기라는 글자가 포함된!
    # args는 인수라는 뜻으로 다른 글자를 사용해도 됨!
    # 그러나 프로그래머들의 관례로 대부분 args로 사용함
    #result = 0 # 결과를 보관하는 변수
    result = ""
    for i in args: # args는 튜플로 들어옴
        result = result + i # 튜플로 되어 있는 값을 번갈아 더함
    return result
```

```python
# add_many 테스트
result = add_many(1,2,3,6)
print(result)
```

```python
resultText = add_many("안녕하세요","방갑습니다.","파이썬 지옥입니다.")
print(resultText)
```

```python
# 함수의 리턴값은 기본 1개이지만 더 만들 수 있다.
def add_and_mul(a, b):
    return a+b, a*b # 결과를 2개로 보냄
```

```python
# 함수에 초기값 지정하기
def say_myself(name, age, man=True):
    #                     man은 기본값으로 true
    print(f"나의 이름은 {name} 입니다.")
    print(f"나의 나이는 {age} 입니다.")
    if man:
        print("저는 남자입니다.")
    else:
        print("저는 여자입니다.")
```

```python
say_myself("김기원","40") # 초기값은 생략해도 자동으로 등록 됨
```

```python
# 변수선언에는 전역변수가 있고 지역변수가 있다.
# 전역변수는 함수나 if, for, while문 바깥쪽! 맨위에 작성하는 것이 좋다.
# 지역변수는 함수나 if, for, while에서 1회용으로 사용되는 것으로 안쪽에
# 선언 한다.
```

```python
g1 = 1
def vartest(g1):
    g1 = g1 + 1
    return g1 # 변수에 리턴을 넣으면 받을 수있는 형태
g1 = vartest(g1) # 함수 실행후 결과를 g1에 다시 넣음
print(g1)
```

```python
a = None   # 전역변수 a
def vartest1(a): # 지역변수 a -> 결론은 서로 다른 a변수
    a = a + 1
    # 리턴 없다.

vartest1(3)
print(a) # a라는 변수는 로컬 변수임으로 def가 종료되면 사라짐.
```

```python
a = None
def vartest2(a):
    a = a + 1
    return a # 위 코드와 변화된 내용은 return과
a= vartest2(3) # 함수 처리후 결과를 a에 넣음
print(a)
```

```python
# 리턴 없이 글로벌 변수를 사용하고 싶은 경우
session = "로그인안함"
def login(id):
    """
    키보드로 입력받은 문자열을
    리스트에서 찾아 중복 검사를 진행하고
    없으면 회원가입을 할 수 있도록 처리

    작성자 : 김기원
    :parameter id : 문자열
    :returns : 로그인 상태를 세션에 저장한다.(글로벌)
    """
    global session # 글로벌 명령어는 전역변수를 가져와!!!!
    session = id   # 글로벌 명령어로 같은 변수명은 사용 금지!!!

login(input("아이디를입력하세요 : "))
print(login.__doc__)
```
